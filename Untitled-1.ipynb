```python
fully_missing = ['V414B', 'V414C', 'V414D', 'V414H', 'V414Q', 'V414R']
df.drop(columns=[c for c in fully_missing if c in df.columns], inplace=True)

print("Dropped 100% missing food columns:", fully_missing)
print("Remaining columns:", len(df.columns))
```

```python
# ─────────────────────────────────────────────────────────────
# STEP 3B — Fix DHS special codes BEFORE any imputation
# ─────────────────────────────────────────────────────────────
def replace_dhs_missing(df):

    # Z-scores: stored as integer * 100, special codes = 9996/9998/9999
    for col in ['HW4', 'HW5', 'HW70']:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')
            df[col] = df[col].replace([9996, 9997, 9998, 9999], np.nan)
            # Only divide if values are clearly in the *100 range
            if df[col].dropna().abs().median() > 50:
                df[col] = df[col] / 100

    # Weight (kg*10) and Height (cm*10)
    for col in ['HW3', 'HW8']:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')
            df[col] = df[col].replace([9996, 9997, 9998, 9999], np.nan)
            if df[col].dropna().median() > 500:
                df[col] = df[col] / 10

    # Haemoglobin (g/dL * 10)
    if 'HW56' in df.columns:
        df['HW56'] = pd.to_numeric(df['HW56'], errors='coerce')
        df['HW56'] = df['HW56'].replace([999], np.nan)
        if df['HW56'].dropna().median() > 50:
            df['HW56'] = df['HW56'] / 10

    # Anaemia: 8/9 = don't know / missing
    if 'HW57' in df.columns:
        df['HW57'] = df['HW57'].replace([8, 9], np.nan)

    # Binary health columns
    for col in ['H11', 'H22', 'H43']:
        if col in df.columns:
            df[col] = df[col].replace([8, 9], np.nan)

    # Remaining food columns
    food_cols = [c for c in df.columns if c.startswith('V414')]
    for col in food_cols:
        df[col] = df[col].replace([8, 9], np.nan)

    return df

df = replace_dhs_missing(df)

# Quick sanity check on z-scores — should be between -6 and +6
print("\nZ-score ranges after fix:")
for col in ['HW4', 'HW5', 'HW70']:
    if col in df.columns:
        print(f"  {col}: min={df[col].min():.2f}  max={df[col].max():.2f}  "
              f"mean={df[col].mean():.2f}")

# Haemoglobin check — should be 5–18 g/dL range
if 'HW56' in df.columns:
    print(f"\n  HW56 (Hb): min={df['HW56'].min():.1f}  "
          f"max={df['HW56'].max():.1f}  mean={df['HW56'].mean():.1f}")
```

```python
# ─────────────────────────────────────────────────────────────
# STEP 4 — Age-aware missing value handling
# ─────────────────────────────────────────────────────────────

# Flag children under 6 months — dietary recall not applicable
df['age_under_6']   = (df['HW1'] < 6).astype(int)
df['age_under_12']  = (df['HW1'] < 12).astype(int)
df['age_under_24']  = (df['HW1'] < 24).astype(int)

# Food columns: fill with 0 for under-6 (not asked), keep NaN for 6+ as missing
food_cols = [c for c in df.columns if c.startswith('V414')]

print(f"\nFood columns available after dropping 100% missing: {food_cols}")

for col in food_cols:
    # Under 6 months: dietary recall not asked → fill with 0
    df.loc[df['HW1'] < 6, col] = df.loc[df['HW1'] < 6, col].fillna(0)
    # 6+ months: remaining NaN → fill with 0 (not consumed)
    df[col] = df[col].fillna(0)

# Deworming (H43): only given 12+ months
# Under 12 months → fill with 0 (not applicable)
# 12+ months → keep 0 as actual "not given"
if 'H43' in df.columns:
    df.loc[df['HW1'] < 12, 'H43'] = df.loc[df['HW1'] < 12, 'H43'].fillna(0)
    df['H43'] = df['H43'].fillna(0)   # remaining 12+ missings → 0

print("\nAfter age-aware fill:")
print(f"  Food cols remaining NaN: {df[food_cols].isnull().sum().sum()}")
print(f"  H43 remaining NaN: {df['H43'].isnull().sum()}")
```

```python
# ─────────────────────────────────────────────────────────────
# STEP 5 — Verify z-scores are valid, flag outliers
# ─────────────────────────────────────────────────────────────

# WHO flags z-scores outside ±6 as biologically implausible
# DHS uses ±5 as the flag limit in practice
for col in ['HW4', 'HW5', 'HW70']:
    if col in df.columns:
        outlier_count = (df[col].abs() > 5).sum()
        df.loc[df[col].abs() > 5, col] = np.nan  # remove implausible values
        print(f"  {col}: removed {outlier_count} implausible z-scores (|z|>5)")
```

```python
# ─────────────────────────────────────────────────────────────
# STEP 6 — Rebuild full missing value summary
# ─────────────────────────────────────────────────────────────

missing2 = df.isnull().sum()
missing2_pct = (missing2 / len(df) * 100).round(2)

final_missing = pd.DataFrame({
    'Column': missing2.index,
    'Missing Count': missing2.values,
    'Missing %': missing2_pct.values
}).query('`Missing Count` > 0').sort_values('Missing %', ascending=False)

print("\nFinal missing value summary:")
print(final_missing.to_string(index=False))
```

```python
import pyreadstat
import pandas as pd
import numpy as np

# Correct way — unpack the tuple
df, meta = pyreadstat.read_sav('IAKR7EFL.SAV')

print(type(df))        # must be <class 'pandas.core.frame.DataFrame'>
print(df.shape)        # must be (232920, ~200+)
print("Loaded ✅")
```

```python
fully_missing = ['V414B','V414C','V414D','V414H','V414Q','V414R']
df.drop(columns=[c for c in fully_missing if c in df.columns], inplace=True)

print(f"Starting shape: {df.shape}")
```

```python
# ── Fix HW5 (WHZ) ────────────────────────────────────────────
# Special code: 9998 = missing, 9980 = flagged/implausible
# Values stored as integer × 100
df['HW5'] = pd.to_numeric(df['HW5'], errors='coerce')
df['HW5'] = df['HW5'].replace([9998, 9980, 9996, 9997, 9999], np.nan)
df['HW5'] = df['HW5'] / 100

# Clip physiologically impossible values (|z| > 6)
df.loc[df['HW5'].abs() > 6, 'HW5'] = np.nan

print(f"HW5 (WHZ) → min={df['HW5'].min():.2f}  max={df['HW5'].max():.2f}  "
      f"median={df['HW5'].median():.2f}  missing={df['HW5'].isna().sum()}")
```

```python
# ── Fix HW70 (HAZ) ───────────────────────────────────────────
df['HW70'] = pd.to_numeric(df['HW70'], errors='coerce')
df['HW70'] = df['HW70'].replace([9998, 9980, 9996, 9997, 9999], np.nan)
df['HW70'] = df['HW70'] / 100

df.loc[df['HW70'].abs() > 6, 'HW70'] = np.nan

print(f"HW70 (HAZ) → min={df['HW70'].min():.2f}  max={df['HW70'].max():.2f}  "
      f"median={df['HW70'].median():.2f}  missing={df['HW70'].isna().sum()}")
```

```python
# ── Drop HW4 (WAZ) — it is not a valid z-score in this file ──
# HW4 contains a different variable (likely percentile rank or
# recode artifact). We will derive WAZ from HW72 flag instead.
# HW72 = 1 means WAZ < -2 (underweight flag) — this is reliable.

if 'HW4' in df.columns:
    df.drop(columns=['HW4'], inplace=True)
    print("HW4 dropped — not a valid z-score in this file version")

# HW72 (underweight flag) is already binary and clean — keep it
print(f"\nHW72 (underweight flag) values:\n{df['HW72'].value_counts(dropna=False)}")
```

```python
# ── Fix HW3 (weight in kg) ───────────────────────────────────
df['HW3'] = pd.to_numeric(df['HW3'], errors='coerce')
df['HW3'] = df['HW3'].replace([9996, 9997, 9998, 9999], np.nan)

# Check if division needed
raw_med_hw3 = df['HW3'].dropna().median()
print(f"HW3 raw median: {raw_med_hw3}")

if raw_med_hw3 > 100:
    df['HW3'] = df['HW3'] / 10   # stored as kg × 10

# Plausible range for under-5: 1.5–30 kg
df.loc[(df['HW3'] < 1.5) | (df['HW3'] > 30), 'HW3'] = np.nan
print(f"HW3 (weight kg) → range=[{df['HW3'].min():.1f}, {df['HW3'].max():.1f}]  "
      f"missing={df['HW3'].isna().sum()}")
```

```python
# ── Fix HW8 (height in cm) ───────────────────────────────────
df['HW8'] = pd.to_numeric(df['HW8'], errors='coerce')
df['HW8'] = df['HW8'].replace([9996, 9997, 9998, 9999], np.nan)

raw_med_hw8 = df['HW8'].dropna().median()
print(f"HW8 raw median: {raw_med_hw8}")

if raw_med_hw8 > 500:
    df['HW8'] = df['HW8'] / 10   # stored as cm × 10

# Plausible range for under-5: 40–130 cm
df.loc[(df['HW8'] < 40) | (df['HW8'] > 130), 'HW8'] = np.nan
print(f"HW8 (height cm) → range=[{df['HW8'].min():.1f}, {df['HW8'].max():.1f}]  "
      f"missing={df['HW8'].isna().sum()}")
```

```python
# ── Fix HW56 (haemoglobin g/dL) ──────────────────────────────
df['HW56'] = pd.to_numeric(df['HW56'], errors='coerce')
df['HW56'] = df['HW56'].replace([999, 9998, 9999], np.nan)

raw_med_hw56 = df['HW56'].dropna().median()
print(f"HW56 raw median: {raw_med_hw56}")

if raw_med_hw56 > 50:
    df['HW56'] = df['HW56'] / 10  # stored as g/dL × 10

# Plausible range: 3–20 g/dL
df.loc[(df['HW56'] < 3) | (df['HW56'] > 20), 'HW56'] = np.nan
print(f"HW56 (Hb g/dL) → range=[{df['HW56'].min():.1f}, {df['HW56'].max():.1f}]  "
      f"missing={df['HW56'].isna().sum()}")
```

```python
# ── Fix HW57 (anaemia classification) ────────────────────────
df['HW57'] = df['HW57'].replace([8, 9], np.nan)
print(f"\nHW57 distribution:\n{df['HW57'].value_counts(dropna=False)}")
# Expected: 0=not anaemic, 1=mild, 2=moderate, 3=severe
```

```python
# ── Fix binary health columns ─────────────────────────────────
for col in ['H11', 'H22', 'H43']:
    if col in df.columns:
        df[col] = df[col].replace([8, 9], np.nan)

print("H11 / H22 / H43 cleaned")
```

```python
# ── Fix food columns ──────────────────────────────────────────
food_cols = [c for c in df.columns if c.startswith('V414')]
for col in food_cols:
    df[col] = df[col].replace([8, 9], np.nan)

print(f"Food columns cleaned: {food_cols}")
```

```python
# ── Age-aware fill for food and deworming ─────────────────────
df['age_under_6']  = (df['HW1'] < 6).astype(int)
df['age_under_12'] = (df['HW1'] < 12).astype(int)

# Food cols: dietary recall not asked for under-6 → fill 0
for col in food_cols:
    df.loc[df['HW1'] < 6, col] = df.loc[df['HW1'] < 6, col].fillna(0)
    df[col] = df[col].fillna(0)  # remaining → 0 (not consumed)

# Deworming: not given under 12 months → fill 0
if 'H43' in df.columns:
    df.loc[df['HW1'] < 12, 'H43'] = df.loc[df['HW1'] < 12, 'H43'].fillna(0)
    df['H43'] = df['H43'].fillna(0)

print("Age-aware fill complete")
```

```python
# ── Impute remaining missing values ───────────────────────────
from sklearn.impute import SimpleImputer

# Median imputation for continuous columns
for col in ['HW3','HW8','HW5','HW70','HW56']:
    if col in df.columns and df[col].isna().sum() > 0:
        df[col] = df[col].fillna(df[col].median())

# Mode imputation for categorical columns
for col in ['HW1','HW2','HW57','HW6','HW7','HW71','HW72',
            'HC61','HC63','V025','H11','H22','V116']:
    if col in df.columns and df[col].isna().sum() > 0:
        df[col] = df[col].fillna(df[col].mode()[0])

print(f"Total missing after imputation: {df.isnull().sum().sum()}")
```
